### Uni-Variant model. only one parameter considered i.e. bedrooms

```python
accom = 3 #we have a house containing 3 bedrooms
file["distance"] = np.abs(file["bedrooms"] - accom) #finding distance b/w all bedrooms and accom
```

```python
file["distance"].value_counts()
```

```python
k = 5 #we will be comparing 5 examples closest to our instance
```

### We are shuffling the data so that there's no biasedness

```python
np.random.seed(1) #seed fixes the random variables. every seed has different set of values.
shuffle = np.random.permutation(file.shape[0])
shuffle #generated a random order of the indexes
```

```python
#we will now set our index equal to shuffle array 
file = file.set_index(shuffle)
file.head()
```

```python
file = file.sort_values("distance")
file.head()
```

```python
file["price"] = file["price"].str.replace("$","").str.replace(",","").astype(float)
```

```python
file["price"]
```

```python
file["price"].head().mean()
```

```python
def k_neighbour(data,bed,k):
    np.random.seed(1)
    shuffle = np.random.permutation(len(data))
    data["distance"] = np.abs(data["bedrooms"] - bed)
    data = data.set_index(shuffle)
    data = data.sort_values("distance")
    k_neighbour = data["price"].head(k).mean()
    return "The average price of {} bedroom apartments is ${}".format(bed,k_neighbour)

                       
```

```python
k_neighbour(file,3,5)
```

```python
k_neighbour(file,4,5)
```

```python
file = pd.read_csv("dc_airbnb.csv")
```

```python
np.random.seed(3)
file = file.set_index(np.random.permutation(len(file)))
file.head()
```

```python
file["price"] = file["price"].str.replace("$","").str.replace(",","").astype(float)
```

### Model will be trained on train data. and validated by applying on test data

```python
train_data = file.iloc[0:int(len(file)*.75)].copy() #slicing and taking 75% of the shuffled data as train data
train_data.shape[0]
```

```python
test_data = file.iloc[2792:].copy() #slicing and taking 25% of the shuffled data as test data
test_data.shape[0]
```

```python
def k_neighbour(bed):
    td = train_data
    np.random.seed(1)
    shuffle = np.random.permutation(len(td))
    td["distance"] = np.abs(td["bedrooms"] - bed)
    td = td.set_index(shuffle)
    td = td.sort_values("distance")
    k_neighbour = td["price"].head().mean()
    return k_neighbour
```

```python
k_neighbour(3)
```

```python
test_data["predicted price"] = test_data["bedrooms"].apply(k_neighbour)
```

```python
test_data[["predicted price","price"]]
```

### This model does a poor job of predicting as there's a lot of difference. This is because a single variable (bedrooms) is considered only. 

* mean absolute error
* mean square error
* root mean square error

#### to avoid negative values we have taken absolute values

```python
mean_abs_error = np.abs(test_data["price"] - test_data["predicted price"]).mean()
mean_abs_error
```

#### to avoid negative values we have squared.

```python
mean_square_error = ((test_data["price"] - test_data["predicted price"])**2).mean()
mean_square_error #unit is also squared so $^2
```

```python
rmse = np.sqrt(mean_square_error)
rmse
```

```python
def k_neighbour(bath):
    td = train_data
    np.random.seed(1)
    shuffle = np.random.permutation(len(td))
    td["distance"] = np.abs(td["bathrooms"] - bath)
    td = td.set_index(shuffle)
    td = td.sort_values("distance")
    k_neighbour = td["price"].head().mean()
    return k_neighbour
```

```python
test_data["predicted price"] = test_data["bathrooms"].apply(k_neighbour)
```

```python
test_data[["predicted price","price"]]
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
feature = "bedrooms"
target = "price"
knn = KNeighborsRegressor(algorithm="brute",metric="euclidean")
knn.fit(train_data[feature],train_data[target])
prediction = knn.predict(test_data[feature])
prediction
```
