```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AmesHousing_1.txt", delimiter= "\t")
#data["Sale Type"].value_counts()
```

```python
null_data = data.isnull().sum()
null_data
```

```python
col_wmv = null_data[null_data==0].index
col_wmv
```

```python
df_no_mv = data[col_wmv]
df_no_mv.head()
```

```python
df_no_mv.isnull().sum()
```

```python
df_no_mv.info()
```

```python
text_cols = df_no_mv.select_dtypes(include=["object"]).columns
text_cols
```

```python
t = df_no_mv["Sale Condition"].value_counts()
t.astype("category")
```

### data type "category" useful if you only have limited string values. can be applied to gender, rating, blood type, social class etc. Addition subtraction operations can't be performed but plots can be obtained.

```python
t = df_no_mv["Sale Condition"].astype("category")
t.value_counts()
```

```python
t.cat.codes.value_counts()
```

```python
df_no_mv["Sale Condition"].head(25)
```

```python
pd.get_dummies(df_no_mv["Sale Condition"]).head()
```

```python
for col in text_cols:
    dummies = pd.get_dummies(df_no_mv[col])
    df_no_mv = pd.concat([df_no_mv,dummies], axis = 1)
    del df_no_mv[col]
```
