```python
import tensorflow as tf
import keras
```

```python
x = file.iloc[:,3:13].values
y = file.iloc[:,13].values
```

```python
#Encoding the categorial data i.e. geography and gender
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le_geo = LabelEncoder()
x[:,1] = le_geo.fit_transform(x[:,1])
le_gender = LabelEncoder()
x[:,2] = le_gender.fit_transform(x[:,2])
ohe = OneHotEncoder(categorical_features=[1]) #creating dummy variable of geography
x = ohe.fit_transform(x).toarray()
x = x[:, 1:]
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from keras.models import Sequential 
from keras.layers import Dense #to add layers
```

```python
#Building the neural network
classifier = Sequential()

#Adding input layer
classifier.add(Dense(6, activation='relu', input_dim = 11))

#Adding hidden layer
classifier.add(Dense(6, activation='relu'))

#Adding output layer
classifier.add(Dense(1, activation='sigmoid'))
```

```python
#Compiling ANN

classifier.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
classifier.fit(x_train, y_train, batch_size=10, nb_epoch=100) #batch size and epochs have no fixed rule for deciding
```
