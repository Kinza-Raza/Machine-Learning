```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
#text = file['text']

ps = PorterStemmer()
corpus = []


for i in range(0,7613):
    text = re.sub(r"https?://[\w\.\w/?[\.]+"," ", file['text'][i]) #removing links
    text = re.sub(r"@\S+", " ", text) #removing tags
    text = re.sub(r"[^\w+#]"," ",text) #removing punctuation except hashtags
    text = text.lower() #lowercase
    text = text.split()
    text = [word for word in text if word not in stopwords.words('english')]
    text = " ".join(text)
    corpus.append(text)
```

```python
file['tweet'] = corpus
file = file.drop(columns='text', axis=1)

file.head()
```

```python
cols = ['tweet','target','label']
file = file[cols]
file.head()
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

file["tokens"] = file["tweet"].apply(tokenizer.tokenize)
file.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer() 
x_train = cv.fit_transform(corpus).toarray() #sparse matrix is created {independent variables}
y_train = file.iloc[:,1].values #dependent variable

cv_1 = CountVectorizer() 
x_test = cv_1.fit_transform(corpus_test).toarray() #sparse matrix is created {independent variables}
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train,y_train)
y_test = nb.predict(x_test)
y_test
```
